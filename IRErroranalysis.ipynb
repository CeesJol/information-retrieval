```python
import pandas as pd
import numpy as np

file_run_results="wikit2019.tsv"
file_relevance_judgements="qrels-pass-binary.txt"

dfres = pd.read_csv(file_run_results, sep="\t", names=["query", "document", "rank"])
dfrel = pd.read_csv(file_relevance_judgements, sep=" ", names=["query", "/", "document", "relevance"])
dfrel = dfrel[["query","document","relevance"]]

```

```python
topRetrieved = 10;
uniqueQueriesRetrievedFor = dfrel['query'].unique();
relRetRows = []
RR = []
for query in uniqueQueriesRetrievedFor:
    relevantDocuments = [];
    irrelevantDocuments = [];
    for i in range(1,topRetrieved+1):
        judgeDocument = dfres.loc[(dfres['query'] == query) & (dfres['rank'] == i)]['document']
        try:
            judgeQD = [query, int(judgeDocument)]
            documentRelevant = (dfrel[['query','document']].values == judgeQD).all(axis=1).any()
            if documentRelevant:
                if (855410 == query):
                    print("found relevant")
                if (len(relevantDocuments) == 0):
                    RR.append(1.0/float(i))
                relevantDocuments.append(int(judgeDocument))
            else: 
                if (855410 == query):
                    print("found irrelevant")
                if ((i == topRetrieved) and (len(relevantDocuments) == 0)):
                    RR.append(0.0)
                irrelevantDocuments.append(int(judgeDocument))
        except:
            RR=RR
    relRetRows.append(dict({'query': query,'relevantCount':len(relevantDocuments),'irrelevantCount':len(irrelevantDocuments),'relevantResults':relevantDocuments,'irrelevantResults':irrelevantDocuments}))
    
relevanceRetrievedCounts = pd.DataFrame(relRetRows,columns=["query", "relevantCount", "irrelevantCount", "relevantResults", "irrelevantResults"])
MRR = np.mean(RR)
```

```python
topRetrieved = 1000;
uniqueQueriesRetrievedFor = dfrel['query'].unique();
relRetRows = []
for query in uniqueQueriesRetrievedFor:
    relevantDocumentsRetrieved = [];
    relevantDocumentsNotRetrieved = [];
    for i in range(1,topRetrieved+1):
        judgeDocument = dfres.loc[(dfres['query'] == query) & (dfres['rank'] == i)]['document']
        try:
            judgeQD = [query, int(judgeDocument)]
            documentRelevant = (dfrel[['query','document']].values == judgeQD).all(axis=1).any()
            if documentRelevant:
                relevantDocumentsRetrieved.append(int(judgeDocument))
        except:
            print("err")
    for document in dfrel.loc[(dfrel['query'] == query)]['document']:
        if int(document) not in relevantDocumentsRetrieved:
            relevantDocumentsNotRetrieved.append(int(document))
            
    relRetRows.append(dict({'query': query,'relevantRetrievedCount':len(relevantDocumentsRetrieved),'relevantNotRetrievedCount':len(relevantDocumentsNotRetrieved),'relevantRetrieved':relevantDocumentsRetrieved,'relevantNotRetrieved':relevantDocumentsNotRetrieved}))
retrievedRelOrNotCounts = pd.DataFrame(relRetRows,columns=["query", "relevantRetrievedCount", "relevantNotRetrievedCount", "relevantRetrieved", "relevantNotRetrieved"])
```
